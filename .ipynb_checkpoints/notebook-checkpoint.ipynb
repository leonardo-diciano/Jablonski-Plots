# Notebook to generate Jablonski diagrams

First input a dictionary with the name of each state and its energy, divided by multiplicity

````python
import matplotlib.pyplot as plt
import numpy as np


singlets={'$S_0$':0.0,'$S_1$':0.800,'$S_2$':1.899,'$S_3$':3.826}
triplets={'$T_1$':1.197}

molecule_name='2TBN-2'
````

````python
sing_pos=[0.2,0.8]
trip_pos=[1.8,2.1]
#trip_pos=[0.9,1.1]

for i in singlets:
    plt.plot( sing_pos,[singlets[i],singlets[i]],'k-',linewidth=2.0)
    plt.text( sing_pos[0]-0.15 , singlets[i] , f"{i}" ,ha='left', va='center', fontsize=10)
for i in triplets:
    plt.plot( trip_pos,[triplets[i],triplets[i]],'k-',linewidth=2.0)
    plt.text( trip_pos[1]+0.15 , triplets[i] , f"{i}" ,ha='right', va='center', fontsize=10)

plt.ylabel("Energy (eV)")
plt.xlim(0,3.5)
plt.xticks([])
plt.yticks(np.arange(0,max(singlets.values())+0.5,0.5))
plt.gca().get_xaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.title(f"State energy diagram\n at MS-CASPT2(12,12)/cc-pVDZ")
plt.show()
````

Input data for the possible decay routes in a tuple with the following structure
```(type, initial state, final state) ```. \
For example, the ISC between S1 and T1 can be written as:
```('ISC','$S_1$','$T_1$')``` .

````python
decay_processes=[
                ('F','$S_1$','$S_0$'), ('IC','$S_1$','$S_0$'),
              ('RISC','$T_1$','$S_0$'),
                ('IC','$S_2$','$S_1$'),
                ('F','$S_2$','$S_0$'),('RISC','$T_1$','$S_1$'),
             # ('ISC','$S_1$','$T_1$'),
                ('F','$S_3$','$S_0$'),
                 ('IC','$S_3$','$S_2$'),
           # ('F','$S_2$','$S_1$'),
            #('ISC','$S_2$','$T_1$') ,#('RISC','$T_1$','$S_2$'),
            #  ('F','$S_3$','$S_1$') ,('F','$S_3$','$S_2$') 
   # ('PH','$T_1$','$S_0$')
]

             
````

````python

for i in singlets:
    plt.plot( sing_pos,[singlets[i],singlets[i]],'k-',linewidth=2.0)
    #if i == '$S_2$':
    #   plt.text( sing_pos[0]-0.12 , singlets[i] + 0.05 , f"{i}" ,ha='left', va='center', fontsize=20)
    #elif i == '$S_3$':
    #    plt.text( sing_pos[0]-0.12 , singlets[i] - 0.05 , f"{i}" ,ha='left', va='center', fontsize=20)
    #else:
    plt.text( sing_pos[0]-0.12 , singlets[i], f"{i}" ,ha='left', va='center', fontsize=20)
for i in triplets:
    plt.plot( trip_pos,[triplets[i],triplets[i]],'k-',linewidth=2.0)
    plt.text( trip_pos[1]+0.12 , triplets[i] , f"{i}" ,ha='right', va='center', fontsize=20)
    
    avoid_list=[]
def draw_wiggly(x,y_start,y_end,dy,color):
    y=np.linspace(y_start,y_end,200)
    x_wiggle=x+0.03*np.sin(2*np.pi*10*(y-y_start)/dy)
    plt.plot(x_wiggle[:-10], y[:-10], color=color,lw=1.0)
    plt.annotate('', xy=(x, y_end), xytext=(x, y[-7]), 
                 arrowprops=dict(arrowstyle='-|>', color=color, lw=3.0,mutation_scale=20 ))

def draw_wiggly_curved(x_start,y_start,x_end,y_end,dy,color):
    t=np.linspace(0,1,300)
    dx=x_end-x_start
    length=np.hypot(dx,dy)
    nx=-dy/length
    ny=dx/length
    if dy < 0 and dx < 0:
        wiggle=0.03*np.sin(2*np.pi*10*t)+(2*t**2)*(1-t)
    elif dy > 0 and dx > 0:
        wiggle=0.03*np.sin(2*np.pi*10*t)+(2*t**2)*(1-t)
    elif dy < 0 and dx > 0:
        wiggle=0.03*np.sin(2*np.pi*10*t)-(2*t**2)*(1-t)
    elif dy > 0 and dx < 0:
         wiggle=0.03*np.sin(2*np.pi*10*t)-(2*t**2)*(1-t)
    angle = np.arctan2(dy, dx)
    cos_a = np.cos(angle)
    sin_a = np.sin(angle)
    x_rot = cos_a * length * t - sin_a * wiggle
    y_rot = sin_a * length * t + cos_a * wiggle
    x_wiggle = x_start + x_rot
    y_wiggle = y_start + y_rot   
    plt.plot(x_wiggle[:-5], y_wiggle[:-5], color=color,lw=1.0)
    plt.annotate('', xy=(x_end, y_end), xytext=(x_wiggle[-5], y_wiggle[-5]), 
                 arrowprops=dict(arrowstyle='-|>', color=color, lw=2.0,mutation_scale=20 ))
points=np.linspace(sing_pos[0],sing_pos[1],len(decay_processes))
colors={'$S_1$':'b','$S_2$':'r','$S_3$':'g','$T_1$':'m'}
for i,j in enumerate(decay_processes):
    if j[0]=='F':
        dy=singlets[j[2]]-singlets[j[1]]
        x=points[i]
        while x in avoid_list:
             x=np.random.uniform(sing_pos[0],sing_pos[1]-0.4)
        avoid_list.append(x)
        plt.arrow(x,singlets[j[1]],0,dy,head_width=0.05,length_includes_head=True,color=colors[j[1]])
        #plt.text(x , singlets[j[1]]+ dy * np.random.uniform(0.05,0.800) , f"$k_r$={j[3]}" ,ha='left', va='center', fontsize=10)
    if j[0]=='IC':
        dy=singlets[j[2]]-singlets[j[1]]
        x=points[i]
        while x in avoid_list:
             x=np.random.uniform(sing_pos[0],sing_pos[1]-0.4)
        avoid_list.append(x)
        draw_wiggly(x,singlets[j[1]],singlets[j[2]],dy,colors[j[1]])
        #plt.text(x +0.03, singlets[j[1]]+ dy * np.random.uniform(0.05,0.800) , f"$k_{{ic}}$={j[3]}" ,ha='left', va='center', fontsize=10)  
    if j[0]=='ISC':
        dy=triplets[j[2]]-singlets[j[1]]
        x_start=sing_pos[1] - np.random.uniform(0.0,0.1)
        while x_start in avoid_list:
            x_start=sing_pos[1] - np.random.uniform(0.0,0.1)
        x_final=trip_pos[0] + np.random.uniform(0.0,0.1)
        while x_final in avoid_list:
            x_final=trip_pos[0] + np.random.uniform(0.0,0.1)
        avoid_list.append(x_start)
        avoid_list.append(x_final)
        if dy > 0:
            draw_wiggly_curved(x_start,singlets[j[1]],x_final,triplets[j[2]],dy,colors[j[1]])
            #plt.text(x_start + (x_final - x_start)/2, max(singlets[j[1]],triplets[j[2]])+ 0.15, f"$k_{{ISC}}$={j[3]}" ,ha='center', va='center', fontsize=10)  
        else:
            draw_wiggly_curved(x_start,singlets[j[1]],x_final,triplets[j[2]],dy,colors[j[1]])
            #plt.text(x_start + (x_final - x_start)/2 -0.05 , singlets[j[1]] + dy/2, f"$k_{{ISC}}$={j[3]}" ,ha='right', va='center', fontsize=10)  
    if j[0]=='RISC':
        dy=singlets[j[2]]-triplets[j[1]]
        x_start=trip_pos[0] + np.random.uniform(0.1,0.2)
        while x_start in avoid_list:
            x_start=trip_pos[0] + np.random.uniform(0.1,0.2)
        x_final=sing_pos[1] - np.random.uniform(0.1,0.2)
        while x_final in avoid_list:
            x_final=sing_pos[1] - np.random.uniform(0.1,0.2)
        avoid_list.append(x_start)
        avoid_list.append(x_final)
        if dy < 0:
            draw_wiggly_curved(x_start,triplets[j[1]],x_final,singlets[j[2]],dy,colors[j[1]])
            #plt.text(x_final +  (x_start - x_final)/2, min(triplets[j[1]],singlets[j[2]])- 0.15 , f"$k_{{rISC}}$={j[3]}" ,ha='center', va='center', fontsize=10)  
        else:
            draw_wiggly_curved(x_start,triplets[j[1]],x_final,singlets[j[2]],dy,colors[j[1]])
            #plt.text(x_final + (x_start - x_final)/2 + 0.20, triplets[j[1]] + dy/2 , f"$k_{{rISC}}$={j[3]}" ,ha='left', va='center', fontsize=10)  
    if j[0]=='PH':
        dy=singlets[j[2]]-triplets[j[1]]
        x=np.random.uniform(trip_pos[0],trip_pos[1])
        plt.arrow(x,triplets[j[1]],0,dy,head_width=0.05,length_includes_head=True,color=colors[j[1]])
        #plt.text(x +0.03, singlets[j[1]]+ dy * np.random.uniform(0.05,0.800) , f"$k_{{ph}}$={j[3]}" ,ha='left', va='center', fontsize=10)  

plt.ylabel("Energy (eV)",fontsize=24)
#plt.xlim(0,2.3)
plt.xlim(0,3.5)
plt.xticks([])
plt.yticks(np.arange(0,max(singlets.values())+0.5,0.5),fontsize=20)
plt.gca().get_xaxis().set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.rcParams["figure.figsize"] = (12, 8)
#plt.title(f"Jablonski diagram for {molecule_name}")
plt.tight_layout()
plt.savefig(f"{molecule_name}.png",transparent=True,format='png',dpi=600)
plt.show()
````
